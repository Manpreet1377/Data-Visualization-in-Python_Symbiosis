# What is the distribution of the Neighbourhood_group?

```python
# Create figure and axis (ax)
fig, ax = plt.subplots(figsize=(8, 6))  # Optional: Adjust the figure size

# Calculate value counts of neighbourhood_group
group_counts = df['neighbourhood_group'].value_counts()

# Plotting the bar chart with counts of each neighbourhood group using ax.bar
bars = ax.bar(group_counts.index, group_counts.values, color='skyblue')

# Adding labels and title
ax.set_xlabel('Neighbourhood Group')
ax.set_ylabel('Count')
ax.set_title('Count of Properties by Neighbourhood Group')

# Adding count numbers on top of each bar
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, yval, ha='center', va='bottom')

# Display the plot
plt.show()
```

# What is the distribution of the room type? Percentage of availability?

```python
# Calculate value counts of room_type
room_type_counts = df['room_type'].value_counts()

# Define labels and sizes for the pie chart
labels = room_type_counts.index
sizes = room_type_counts.values

# Plotting the pie chart
plt.figure(figsize=(4,4))  # Optional: Adjust the figure size
plt.pie(sizes, labels=labels, explode=(0.1, 0, 0), shadow=True, autopct='%1.1f%%', startangle=360)

# Adding a title


# Display the plot
plt.show()
```

# Frequency of the Number of reviews?

```python
# Extract the 'number_of_reviews' column
number_of_reviews = df['number_of_reviews']


# Plotting the histogram
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
# Plot the histogram
n, bins, patches = plt.hist(number_of_reviews, bins=10, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Reviews')

# Adding count values on top of each bar
for i in range(len(patches)):
    x = bins[i] + (bins[i + 1] - bins[i]) / 2  # Calculate x-coordinate for placing text at the center of the bar
    y = n[i]  # Get the height of the bar (frequency)
    plt.text(x, y + 0.5, int(y), ha='center', va='bottom')

# Display the plot
plt.show()
```

# Is there any correlation between number of reviews and the price?

```python
# Plotting the scatter plot
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
plt.scatter(df['number_of_reviews'], df['price'], color='skyblue')

# Adding labels and title
plt.xlabel('Number of reviews')
plt.ylabel('Price')
plt.title('Scatter Plot: Reviews vs Price')

# Display the plot
plt.show()
```
